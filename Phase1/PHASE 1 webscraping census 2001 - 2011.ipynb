# Importing Libraries for Webscrapping

```python
# importing necessary libraries
import pandas as pd
import selenium
from selenium import webdriver
from selenium.webdriver.common.by import By
import time
```

# Area 2001

```python
# opening chrome via webdriver
driver=webdriver.Chrome()
```

```python
# getting link of area
driver.get('https://en.wikipedia.org/wiki/List_of_states_and_union_territories_of_India_by_area')
```

```python
# Adding COlumns Name
state_name=[]
area=[]
region=[]
```

```python
state=driver.find_elements(By.XPATH,'/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[2]/tbody/tr/td[2]/a')
for i in state:
    state_n=i.text
    state_name.append(state_n)

areaa=driver.find_elements(By.XPATH,'/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[2]/tbody/tr/td[3]')
for i in areaa:
    ar=i.text
    area.append(ar)
    area=area[0:36]

reg=driver.find_elements(By.XPATH,'/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[2]/tbody/tr/td[4]/a')
for i in reg:
    rg=i.text
    region.append(rg)
```

```python
# Making dataframe of collected data
area_of_state=pd.DataFrame({'State':state_name,'Area':area,'Region':region})
```

```python
area_of_state
```

```python
area_of_state.to_csv('area_of_state.csv')
```

# Census 2001

```python
driver=webdriver.Chrome()
```

```python
driver.get('http://cyberjournalist.org.in/census/cenindia.html')
```

```python
State=[]
population=[]
male_popu=[]
female_popu=[]
```

```python
states=driver.find_elements(By.XPATH,'/html/body/center[2]/table/tbody/tr/td[2]/font/font')
for i in states:
    stat=i.text
    State.append(stat)
```

```python
popul=driver.find_elements(By.XPATH,'/html/body/center[2]/table/tbody/tr/td[3]/font/font')
for i in popul:
    pop=i.text
    population.append(pop)
```

```python
male=driver.find_elements(By.XPATH,'/html/body/center[2]/table/tbody/tr/td[4]/font/font')
for i in male:
    malee=i.text
    male_popu.append(malee)
```

```python
female=driver.find_elements(By.XPATH,'/html/body/center/table/tbody/tr/td[5]/font/font')
for i in female:
    fem=i.text
    female_popu.append(fem)
```

```python
len(female_popu)
```

```python
# Making dataframe of collected data
population=pd.DataFrame({'State':State,'Population':population,'Male':
                        male_popu,'Female':female_popu})
```

```python
population
```

```python
population.to_csv('population.csv')
```

# Census 2011

```python
driver=webdriver.Chrome()
```

```python
# Census 2011
State=[]
Total_Population=[]
Males=[]
Females=[]
Literacy=[]
Area=[]
Decade_Growth=[]
```

```python
driver.get('https://en.wikipedia.org/wiki/2011_census_of_India')
```

```python
state=driver.find_elements(By.XPATH,'/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[6]/tbody/tr/td[2]/a')
for i in state:
    stat=i.text
    State.append(stat)

total=driver.find_elements(By.XPATH,'/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[6]/tbody/tr/td[5]')
for i in total:
    tot=i.text
    Total_Population.append(tot)

male=driver.find_elements(By.XPATH,'/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[6]/tbody/tr/td[7]')
for i in male:
    males=i.text
    Males.append(males)

female=driver.find_elements(By.XPATH,'/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[6]/tbody/tr/td[8]')
for i in female:
    fem=i.text
    Females.append(fem)

literacy=driver.find_elements(By.XPATH,'/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[6]/tbody/tr/td[10]/div/div[2]')
for i in literacy:
    liter=i.text
    Literacy.append(liter)

area=driver.find_elements(By.XPATH,'/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[6]/tbody/tr/td[13]')
for i in area:
    areaa=i.text
    Area.append(areaa)

decade=driver.find_elements(By.XPATH,'/html/body/div[2]/div/div[3]/main/div[3]/div[3]/div[1]/table[6]/tbody/tr/td[15]')
for i in decade:
    decc=i.text
    Decade_Growth.append(decc)
```

```python
# Making dataframe of collected data
census2011=pd.DataFrame({'State':State,'Total Population':Total_Population,'Males':Males,'Females':Females,
                         'Literacy':Literacy,'Area':Area,'Decade_Growth 2001-11':Decade_Growth})
```

```python
census2011.to_csv('census2011.csv')
```

# Literycay Rate 2001

```python
driver=webdriver.Chrome()
```

```python
driver.get('http://cyberjournalist.org.in/census/cenlit0.html')
```

```python
State_litercay2001=[]
Total_litercay2001=[]
Males_Literacy2001=[]
Females_Literacy2001=[]
```

```python
state=driver.find_elements(By.XPATH,'/html/body/center[2]/table/tbody/tr/td[2]/font/font')
for i in state:
    stat=i.text
    State_litercay2001.append(stat)

total=driver.find_elements(By.XPATH,'/html/body/center[2]/table/tbody/tr/td[3]/font/font')
for i in total:
    totall=i.text
    Total_litercay2001.append(totall)

males=driver.find_elements(By.XPATH,'/html/body/center[2]/table/tbody/tr/td[4]/font/font')
for i in males:
    male=i.text
    Males_Literacy2001.append(male)

females=driver.find_elements(By.XPATH,'/html/body/center[2]/table/tbody/tr/td[5]/font/font')
for i in females:
    female=i.text
    Females_Literacy2001.append(female)
```

```python
# Making dataframe of collected data
Literacy_Rate_2001=pd.DataFrame({'State':State_litercay2001,'Total_litercay2001':Total_litercay2001,'Males_Literacy2001':Males_Literacy2001,
                                'Females_Literacy2001':Females_Literacy2001})
```

```python
Literacy_Rate_2001.to_csv('Literacy_Rate_2001.csv')
```
