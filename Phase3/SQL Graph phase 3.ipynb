```python
import pandas as pd 
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import warnings 
warnings.filterwarnings('ignore')
```

```python
ipc=pd.read_csv('murder rape ipc crimes.csv')
```

```python
ipc
```

```python
plt.figure(figsize=(15,8),facecolor='powderblue')
sns.lineplot(data=ipc, x='state_ut',y='MAX(murder)',hue='year',marker='o',linestyle='--',palette='husl')
plt.xlabel('STATE_UT')
plt.ylabel('MURDER')
plt.title('In WHICH STATE MAX MURDER HAPPEN')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,8),facecolor='powderblue')
sns.lineplot(data=ipc,x='year',y='MAX(murder)',marker='o',linestyle='--',palette='viridis')
plt.xlabel('Year')
plt.ylabel('MURDER')
plt.title('MAX MURDER YEAR')
plt.show()
```

```python
plt.figure(figsize=(20,8),facecolor='powderblue')
sns.lineplot(data=ipc,x='district',y='MAX(murder)',hue='year',palette='viridis')
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(15,8),facecolor='powderblue')
sns.lmplot(data=ipc,x='year',y='MAX(murder)',hue='state_ut',aspect=15/10)
plt.show()
```
